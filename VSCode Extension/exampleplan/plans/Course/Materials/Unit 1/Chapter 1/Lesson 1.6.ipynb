# Lesson 1.6: Dictionaries 

Dictionaries in python are exactly like dictionaries in real life!
They store values in a key: value pair. 

Think of it like dictionaries with the word: definition. 
The word is the key that you look up in the dictionary and the value is the definition that you find. 

Like dictionaries in real life, **duplicate values are not allowed** but the dictionary is **changeable** as needed. 

How is this different than a list? 
No duplicates!

Initializing a dictionary:
To set values in your dictionary, declare a variable and then assign it to {}. Within the brackets define your key: value pairs, separating each one with a comma. 
Keys can be strings or int's but in most cases it makes sense to have string keys. Values can also be any datatype. 

To define a dictionary follow this example:

```python
date = { "month": "Novemeber", "day": "Thursday", "year": 2025 }
print(date)
```

You can also print specific values using the keys, like `print(date["year"])`

## Methods
Dictionary methods are a little different than lists since we are now working with key: value pairs. 

Here are a few cool methods to use on dictionaries!
Remember you can call a method on a dictionary by using the variable name created for the dictionary . the method, like this `dict_name.method()`

1. keys() - returns a list of the dictionary's keys
2. values() - returns a list of all the values in the dictionary
3. get(key) - returns the value associated with that key in the dictionary 
4. update({key: value}) - adds the key: value pair to the dictionary
5. pop(key) - removes the key: value pair connected to this key
6. clear() - removes all elements from the dictionary
7. copy() - returns a copy of the dictionary you can store in another variable 
8. len(dictionary_name) - returns number of key: value **pairs** in the dictionary
