# Purpose of prints
Printing is an important way of displaying information to your console.
It’s perfect for showcasing what you’ve accomplished or to debug faulty code.

**Commands** are instructions we give to the computer to tell it what to do.

To print you use the `print()` command.

Inside the parentheses include the information you want to print. This can be of any data type. A few examples, 
- String: “hello!”
- Variable: rover_name
- Int: 18
- Float: 2.3

### Try it out!

```python
print("Coding is fun!")
```

## Combining strings and variables in print statements
In case you want to include the value of a variable within a sentence in your print statement, you can use `print(f “{}”)`. This **command** formats your sentence. Inside the quotes type in your text and any variables go in brackets {}.

For example, 

```python
rover_name = "Courage"
print(f"The name of my favorite rover is {rover_name} and its super cool!")
```

Try running this code! What does it print? 

#### Side Note - Printing on a **new line**
Sometimes we want to use one print statement but print multiple lines of text. 
To do this use the escape sequence `\n` within your string.
An escape sequence is designated by the \ symbol, the letter n lets the computer know you want everything after the escape sequence to be on a new line. 
Since this is an 'escape' sequence we want there to be something for it to escape from! Always make sure you only use escape sequence's inside a string. This means they have to be in quotes for them to work.

Try it out!

```python
print("This is my first line \nThis is my second line!")
```

You can also concatenate `\n` to existing prints like this 

```python
print("line one" + "\n" + "line two")
```

Why do we think this works?
It's because escape sequences only work in strings and you can concatenate strings together!

### Bugs and Issues
Make sure the \ in \n is pointing left to right, instead of this / which points right to left.
For the escape sequence to work it has to be \n not /n.
