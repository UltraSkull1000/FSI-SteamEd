# Lesson 1.1: Variables

A variable is a name used to reference specific pieces of data.

With variables you can quickly call and use different pieces of information without having to explicitly list out everything they contain. 

For example: `name = “Courage The Rover”`, where ‘`name`’ is your variable and `“Courage The Rover”` is the value being assigned to the variable.

To name a variable there are a few important conventions to follow:  
1. Always start with a letter or _

2. Can only contain letters, numbers, and _

3. Case sensitive so watch out for those upper and lower case letter differences!

4. Avoid using Python keywords or function names (don’t worry we’ll go over specifics in another lesson)

Once you make a variable name, use assignment to give it a value. 

Assignment is when you set the variable equal to a value with the = symbol.

Just like when we did `name = “Courage The Rover”`

## Bugs and Issues:

Make sure the variable equals an appropriate data type. For example, if your variable holds words, that makes the data type a string and the value has to be in quotes. For more information on data types check out Lesson 1.2!
