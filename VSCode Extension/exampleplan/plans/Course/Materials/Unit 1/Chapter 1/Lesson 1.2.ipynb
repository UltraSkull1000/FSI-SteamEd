# Lesson 1.2: Data Types

The computer stores information in different data types. Think of it as the difference between numbers, words, and true/false.

## Numbers
For numbers, use the int or float data types.

### Integers
- int represents integers.
- These are whole numbers like 1, 2, and 3. 
- For example, `integer = 7`

### Decimals
- float represents decimal numbers like 3.7
- For example, `floating = 3.7`

## True or False?

### Boolean
For true/false values, use the bool data type. This stands for Boolean.

The `bool` data type outputs true or false, otherwise known as 1 (true) or 0 (false) values. This is really helpful for conditional statements discussed in later lessons!

- For example, `true_or_false = 1`

## Words for my stories
For words, use the String data type.

Words are stored in strings. Strings are a sequence of characters (letters, numbers, spaces, or special characters like !@#$%&) that **have to be in quotes**.

For example, `“Courage The Rover”` is a string because it is in quotes. 

The cool thing is the type of quote doesn't matter, it can be a single or double quote, but whatever you start with has to be the same kind of quote you close the string with. 
`'Python is fun'` is also a string because its surrounded by single quotes. 
`Am I a String` is not a string because its not surrounded by quotes. 
The same with `(is this a string?)` because it is not surrounded by quotes. 

What about `“String?'` , that's a tricky one! Remember the quotes have to start and end as the same kind. 

Fun Fact! Python is dynamically typed. This means Python checks the types of your variables during run time so you don’t have to explicitly declare the data type. This gives you flexibility to change the values of your variables as you go!
