# Lesson 2.1: So what if?

The first conditional we are going to learn is the if statement!

Have you ever wondered "what if this happens?"
This is the basics for building if statements.

If statements follow a if (something): "do this" structure 
For example, 

```python
x = 3
if x > 2:
	print(f"The number {x} is greater than 2!")
```

Key things to note here:

- Always put a colon : at the end of the condition for the if statement. 

- the "do this" portion of the code has to be **indented** below the if statement. This makes it easier to read and tells the computer what is part of the if statement.

What if the condition given to the if statement is false? 

If you have a different task for the code to do if the condition is false, you can put it under an if-else statement. 

The else statement follows the same structure as if statements, with a : after the condition and the content being indented. 

This follows a 

```python
if (condition):
    # do this
else:
    # do this
```

One more case. What if we have multiple if else statements we want to put together? Then we can use elif.

elif is the combination of else if. This follows a 

```python
if(condition): 
	# do this
elif(condition):
	# do this
# *how many ever elif condition statements you want*
else:
	# do this
```

**Try it out!**

```python
grade = 85
if grade > 90:
	print("A!")
elif grade > 80:
	print("B!")
else:
	print("C!")
```

Did you know you can compare Strings too? The key is to use a data type that is comparable using < > <= >= = or == (== is used to compare equality of strings)

#### Side Note - **Pseudocode**
Pseudocode is your design notes for the code you write.

They are a high level description of each line in your code, and are written in plan English. 

The goal is to capture the logic of your **algorithm** (process/task).

A few rules with pseudocode:

1. Use clear and concise language, remember high level! You should be able to understand the basics at a glance. Also, keep your language consistent to help readability. 

2. Include basic programming structures like if statements, or for and while loops.

3. Make sure to indent correctly. While we won't be running pseudocode, it helps organize your thoughts and matches the basic structure of the coding concepts you will be translating to code. 

For example, the code from the example above would look like this in pseudocode:

```python
# set grade to 85
# if grade is greater than 90, then
    # print A!
# else if grade is greater than 80, then
    # print B!
# else
    # print C!
```
