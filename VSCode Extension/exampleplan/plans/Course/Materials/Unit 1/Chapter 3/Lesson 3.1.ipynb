# Lesson 3.1: Declaration

Before you use a function you have to define it!
To define a function there are 3 steps.

**Step 1**: pick a function name using the same conventions as variable names (start with a letter, only use letters, numbers or underscore), and make sure the name matches the task you are trying to do!

**Step 2**: pick your parameters 

**Step 3**: put it all together following this structure:

	`def function_name(parameter1, parameter2, parameter3...):`

The declaration starts with def because we are defining a new function. The number of parameters you have is dependent on what you are trying to do. 

Which brings us to the question, what is a parameter?

A parameter is variable that will hold information you pass to function. The information you choose to pass along are things the function needs to do its task.

For example, if you want to add two numbers then the function parameters would be the two numbers you are adding

```python
def add_two_numbers(number1, number2):
	print(number1 + number2)
```
