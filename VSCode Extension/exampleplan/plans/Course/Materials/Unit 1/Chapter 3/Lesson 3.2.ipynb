Content for Python functions is the code for what you're trying to accomplish. 

The most important part to remember here is that the content of your function must be **indented** in so that everything is under your function declaration. 

VSCode will automatically do this for you as long as you have a : at the end of the function definition and you've hit **Enter** (on your keyboard) after.

For example, 

```python
def where_to_go(location, distance):
	rover_name = "Courage"
	if(location == "far"):
		print(f"The distance {distance} miles is too far for {rover_name}!")
	else:
		print(f"{rover_name} is traveling to {location}!")
```
