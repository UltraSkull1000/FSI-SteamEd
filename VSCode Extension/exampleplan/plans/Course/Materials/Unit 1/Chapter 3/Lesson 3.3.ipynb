Let's create a main method to call our function. A main method is where most programs will start running their code. 

Python doesn't have a dedicated main method but it does have a few special variables we can use to create our main method.

Try it out!

```python
def main():
	print("In main!")
	
if __name__ == "__main__":
	main()
```

Now lets include our function with this code.

```python
def main():
	print("In main!")
	
def where_to_go(location, distance):
	rover_name = "Courage"
	if(location == "far"):
		print(f"The distance {distance} miles is too far for {rover_name}!")
	else:
		print(f"{rover_name} is traveling to {location}!")
	
if __name__ == "__main__":
	main()
```

Now to call our function in the main method!

To call a function, type in the function name and in parentheses any parameters it might have. If there are no parameters, then use empty parentheses. This should match your function definition, ex. `function_name()`

Make sure to declare any variables you are planning to use in main and initialize them to a value, before calling the function in main.

Try it out!

```python
def main():
	print("In main!")
	distance = 100
	if(distance > 50):
		location = "far"
	else:
		location = "space station"
	where_to_go(location, distance)
	
def where_to_go(location, distance):
	rover_name = "Courage"
	if(location == "far"):
		print(f"The distance {distance} miles is too far for {rover_name}!")
	else:
		print(f"{rover_name} is traveling to the {location}!")
	
if __name__ == "__main__":
	main()
```

## Return
What if we want to return a value back to our main method from our function? 

We can use a return statement to do this!

A return statement looks like this, `return value`.

It will return whatever variable value you have designated in your function to the function or method that called it. 

Return statements will **always** be the **last line** your code will compute in a function because it will leave the function to return the value. 

You can have more than one return statement if they are in conditional statements. For example, when you use if else statements you can use one return for the if and one return for the else portion of the code. 

Try it out!

```python
def add_two_numbers(number1, number2):
	answer = number1 + number2
	return answer

def main():
	print("In main!")
	number1 = 5
	number2 = 2
	print(add_two_numbers(number1, number2))
	
if __name__ == "__main__":
	main()
```

## Bugs and Issues
If you receive an error, the first part of bug fixing is to read the error message. 

Here are a few common errors and what to check

1. **SyntaxError**: Syntax is the rules for writing code. Normally this means you are missing a colon : at the end of definitions or conditionals, or you are missing quotes "" or parentheses in the appropriate places. Double check your code to make sure none of this is missing. 

2. **IndentationError**: Remember that the content of a function always has to be indented in. Make sure you have a colon at the end of your function definition and press Enter for Python to automatically indent or tab over the necessary code yourself.

3. **NameError**: Normally means you spelled something incorrectly. Double check the spelling of all your commands. Does your code say prnt instead of print? Or whiel instead of while? Correct those and you'll be good to go!
