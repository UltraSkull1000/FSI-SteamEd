## Time for math class
Operators in coding are very similar to operators in math but with a few special additions.

### Assignment 
Assignment is setting equal to something. It essentially stands for the = symbol. 
Assignment is used to compute math problems or store content in variables.
### Addition
Addition in coding is exactly the same in math. If you want to add 2 numbers use the + operator. If you want to store the answer you can set the operation = a variable, or you can print the answer out directly.

For example,

```python
answer = 4 + 5
print(answer)
```

or, 

```python
print(5+11)
```

#### Side Note - **Comments**:
Comments are a way to add notes and thoughts to your code. These comments will be ignored when running the program so there are no conventions beyond letting the computer know there is a comment!

To start writing a single-line comment use the `#`
- For example, `# This is a comment.`
- In larger code snippets, comments can come before the line or section they explain or on the same line

```python
print("Operators in coding") # These are the operators + - * / 
# comments explaining code are helpful to quickly understand what a section does
```

### Subtraction
Subtraction is the same as in math too! To subtract 2 values use the - symbol.
Storing the answer is the same for all operators. You can assign the value to variables and print or print directly. 

**Try it out!** For the example below, add a comment above the print line of what you think will print, then run the cell to test if you were right!

```python
operand1 = 3
operand2 = 7
answer = operand2 - operand1
print(answer)
```

### Multiplication & Division
Multiplication and division are also the same as in math. The symbol to do multiplication is going to be `*`. This is the symbol traditionally above the number 8 on the keyboard, press Shift + 8 to get the multiplication symbol. The symbol to do division is `/`

Math operations on two int's will always result in an int. If one of the values are a floating point (float) value then the answer will also be a float (have decimals).

**Try it out!** For the example below, add a comment above the print line of what you think will print, then run the cell to test if you were right!

```python
answer = 4 * 5
print(answer)
```

What about this one?

```python
answer = 7.0 * 3.0
print(answer)
```

One last one!

```python
answer = 12 / 2.0
print(answer)
```

### Concatenation: When did we start adding words?
Concatenation is the process of joining multiple strings together to make one, new string. The thought of adding words together might be confusing but its actually really simple! 
When you add strings together you are attaching them to the end of each other. 
For example String1 + String2 would create the new string String1String2.

**Try it out!**

```python
firstName = "Courage"
lastName = "the Rover"
# What will the following statement print? Answer here before running code: 
print(firstName + lastName)
```

Hmmm. Something is a little off here. Our words are all smushed together and hard to read! To fix this issue, you can add a string with a blank space in between firstName and lastName like this `" "`. Or you can put a space before the world 'the' in the string stored in variable lastName.

Remember, a string can be any sequence of characters and a blank space counts as a character!
