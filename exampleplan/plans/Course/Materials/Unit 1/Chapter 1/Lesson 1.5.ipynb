# Lesson 1.5: Lists

Lists are a way to store multiple values in one variable.
Think of it like a grocery list with multiple items on it.

For example, 

```python
grocery_list = ["apples", "cereal", "milk", "chocolate"]
print(grocery_list)
```

What are some important differences you see between this and a normal variable declaration?

The brackets [] are an important distinction! Brackets let the computer know we are starting a list. 

Commas separate the entries in our list. Entries can be of different data types but it is recommended to keep lists to one datatype, such as all strings or all int. This makes our lists more readable and easy to use. 
Imagine you were using a list called `names` and suddenly found a number. It would be confusing trying to figure out what the number means exactly. 

## Methods
Three key things to know about lists is that they are **ordered**, allows **duplicate values**, and is **changeable**.

### Ordered
To keep things organized, lists have a specific order to them. 
Think of it like numbering your grocery list. 

This is called indexing. Indexing starts with the number 0 instead of 1. This means the last item on your list will always be indexed one number less than the number of items you have

A visual example will help:
`list = ["robots", "space", "rockets", "moon"]`
	0. robots
	1. space
	2. rockets
	3. moon
There are 4 items total in the list. 
Item 1 is robots and it is in **index** 0.
Item 2 is space and it is in index 1.
Item 3 is rockets and it is in index 2.
Item 4 is moon and it is in index 3.

Do we see how there are 4 items but the indexes only go till 3? 
This is because we started at 0!

To make this less confusing we won't use the item number of the list and instead will only use the **index** number.

### Allows duplicates
Since our lists are ordered, having duplicate values are okay because we will always be able to tell which index each value is in.

For example, 
`list = ["rovers", "robots", "space", "lab", "rovers"]`

### Changeable
Lists are also changeable, meaning there are lots of really cool ways to add, change, and remove items from a list.
These different actions you can do to a list are called methods. 
To use a method the code is `list.method()` where method() is the specific name of the method you are trying to call.

Here are a few cool methods to try out!
1. append(item) - adds the item to the end of the list.
2. insert(index, item) - inserts the item that you want at a specific index. This changes the order of your list so make sure that you count indexes from 0 before inserting!
3. remove(item) - removes the first instance of that item in your list  
4. clear() - removes all items from the list
5. copy() - returns a copy of the list you can store in another list
6. sort() - sorts the list in alphabetical or numerical order

In case you wanted to know the number of items in a list (the actual number not our indexing), you can use the len method.
For example, 

```python
list = ["robots", "space", "rockets", "moon"]
print(len(list))
```

will print 4, because there are 4 items in the list.
